**ทำการ import library ที่จำเป็น**

```python
from numpy import argmax
from pandas import read_csv
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
import tensorflow as tf
```

**ทำการโหลดข้อมูล**

```python
#load the dataset
path = '/content/drive/MyDrive/1204303 - Artificial Intelligence/Apple Quality Classification/dataset/apple_quality.csv'
df = read_csv(path)
```

```python
df
```

**ทำการ drop คอลัมน์ A_id ทิ้งไปเนื่องจากเราไม่ใช้งานเพราะมันเป็นเเค่ index**

```python
df.drop(['A_id'], axis=1, inplace=True)
df
```

**ทำการเเบ่งข้อมูล**

```python
# split into input and output columns
X, y = df.values[:, :-1], df.values[:, -1]

# ensure all data are floating point values
X = X.astype('float32')

# encode strings to integer
y = LabelEncoder().fit_transform(y)

# split into train and test dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

**ทำการพิจารณาจำนวนของ Input**

```python
# determine the number of input features
n_features = X_train.shape[1]
print(n_features)
```

**ทำการสร้าง Model**

```python
# define model
model = Sequential()
model.add(Dense(10, activation='relu', kernel_initializer='he_normal',
                input_shape=(n_features, )))
model.add(Dense(8, activation='relu', kernel_initializer='he_normal'))
model.add(Dense(1, activation='sigmoid'))
```

**ทำการกำหนดการประมวลผล**

```python
# compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

**ทำการ Train model**

```python
# fit the model
model.fit(X_train, y_train, epochs=150, batch_size=32, verbose=1)
```

**ทำการประเมิน Model**

```python
# evaluate the model
loss, acc = model.evaluate(X_test, y_test, verbose=0)
print(f'Test Accuracy: {acc:.3f}')
```
